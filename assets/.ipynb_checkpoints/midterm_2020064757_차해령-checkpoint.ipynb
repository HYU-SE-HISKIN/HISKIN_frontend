```python
#task1 (1) : import modules and read "2019-07-trade.csv"

import pandas as pd
import numpy as np

df1=pd.read_csv("/Users/infosys/Desktop/ohhoho/AI/midterm/2019-07-trade.csv")
df1.head()
```

```python
#task1 (2) : create new column 'signed_quantity' and this make the value of quantity (+) if the side is 0(buy) else (-) the side is 1(sell)

df1['signed_quantity'] = df1.apply(lambda row: row['quantity'] if row['side'] == 0 else -row['quantity'], axis=1)
```

```python
#task1 (3) : create new column 'quantity_cumulative' which means the cumulative values of signed_quantity

df1['quantity_cumulative'] = df1['signed_quantity'].cumsum()
df1.head(20)
```

```python
#task1 (4) : If accumulative quantity is close to 0 (less than 0.0001), represent the value as 'df_task1' dataframe.
            #We can know the moment that the number of quantities bought and sold are equal is index 19289 (2019-07-26 01:46)

df_task1=df1.loc[(abs(df1['quantity_cumulative']) < 0.0001)]
df_task1
```

```python
#task1 (5) : get sum of the number of quantities both bought and sold from index 0 to index 19289

df_buy_sum=df1['amount'][:19290].apply(lambda x : x if x<0 else 0).sum()
df_sell_sum=df1['amount'][:19290].apply(lambda x : x if x>0 else 0).sum()

df_buy_sum, df_sell_sum
```

```python
#task1 (6) : calculate profit with the sum of the quatities bought and sold

profit = df_buy_sum + df_sell_sum
profit
```

```python
#task2 (1)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df2=pd.read_csv("/Users/infosys/Desktop/ohhoho/AI/midterm/2019-07-trade.csv")
df2.head()
```

```python
#task2 (2)


df2[['date', 'time']] = df2['timestamp'].str.split(expand=True)
dates=df2['date'].str.split('-')
df2['day']=dates.str.get(2)

df2.sample(10)
```

```python
#task2 (3)

grouped=df2.groupby(['side','day']).size().unstack()
grouped.fillna(0, inplace=True)

grouped
```

```python
#task2 (5)

grouped=grouped.transpose()
grouped.plot(title='Daily Transaction Count', xlabel='timestamp_day')
grouped.show()
```
